# Regresión Lineal Simple

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

## 1. Preparación de Datos

```python
df = pd.read_csv("./Advertising.csv")
df.head()
```

```python
df.describe()
```

```python
sns.pairplot(df)
```

## Variables a evaluar

```python
x = df[["TV"]]
y = df[["sales"]]
```

## Separación de conjunto de entrenamiento y prueba

```python
%pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
y_prediction = model.predict(x_test)
y_prediction
```

```python
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
print(f"Y axis interceptor: {model.intercept_}")
print(f"Coeficient: {model.coef_}")
```

```python
print(f"Error de media: {mean_absolute_error(y_true=y_test, y_pred=y_prediction)}")
print(f"Error Cuadratico medio: {r2_score(y_true=y_test, y_pred=y_prediction)}")
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x=x_test, y=y_test, color="blue")
plt.plot(x_test, y_prediction)
plt.xlabel("Tv")
plt.ylabel("Sales")
plt.title("Regresion lineal simple")
```
