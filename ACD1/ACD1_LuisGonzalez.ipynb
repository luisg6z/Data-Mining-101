# Limpieza de datos

```python
new_df = df.drop(labels=["PassengerId", "Name", "Ticket", "Cabin"], axis=1)
new_df.head()
```

## Valores nulos en el campo "Age"

```python
new_df["Age"].isnull().sum()
```

86 valores nulos

## Reemplazando valores nulos en el dataframe

```python
age_mean = new_df["Age"].median()
new_df["Age"].fillna(age_mean, inplace=True)
new_df["Age"].isnull().sum()
```

## Valores nulos en el campo "Embarked"

```python
new_df["Embarked"].isnull().sum()
```

## Moda de Embarked

```python
new_df["Embarked"].mode()
```

# Filtrado y Transformación

```python
import numpy as np
```

## Crear columna family_size

```python
new_df["family_size"] = new_df["SibSp"] + new_df["Parch"]
new_df["family_size"].head()
```

```python
sex_mode = new_df["Sex"].mode()[0]
new_df["Sex"].fillna(sex_mode, inplace=True)
new_df["Sex"].isnull().sum()
```

## Filtrar los menores de edad

```python
under_age_df = new_df[new_df["Age"] < 18]
under_age_df.head()
```

## Cambiar el tipo de dato de la columna de sexo

```python
under_age_df["Sex"].isnull().count()
```

Acá se rellenan los datos nulos de la columna de sexo por la moda de esta columna

```python
sex_mode = under_age_df["Sex"].mode()[0]
under_age_df["Sex"].fillna(sex_mode, inplace=True)
under_age_df["Sex"].isnull().sum()
```

```python
under_age_df["Sex"] = under_age_df["Sex"].map({'male': 1, 'female': 0})
under_age_df["Sex"] = under_age_df["Sex"].astype(np.int8)
under_age_df["Sex"].head()
```

## Total de menores de edad en el Titanic

```python
under_age_rows, under_age_cols = under_age_df.shape
print(f"rows: {under_age_rows}")
```

Habia un total de 41 menores de edad en el Titanic

# Agrupamiento y Estadistica

## Edad promedio por Sexo

```python
new_df.groupby("Sex")["Age"].mean()
```

La mayor edad promedio la tienen las mujeres con un promedio de 29.73 años

## Edad promedio por clase

```python
new_df.groupby("Pclass")["Age"].mean()
```

La mayor edad promedio la tienen los pasajeros de 1era clase con un promedio de 39.75 años

## Tasa de supervivencia por puerta de embarque

```python
new_df.groupby("Embarked")["Survived"].mean() * 100
```

Los pasajeros con mayor tasa de supervivencia son los que embarcaron en el puerto Q con un 52.17% de supervivencia

## Tasas de supervivencia por edad

```python
minors = new_df[new_df["Age"] < 18]
minors_survival_rate = minors["Survived"].mean() * 100
print(f"Minors survival rate: {minors_survival_rate:.2f}%")
```

```python
adults = new_df[(new_df["Age"] >= 18) & (new_df["Age"] <= 50)]
adults_survival_rate = adults["Survived"].mean() * 100
print(f"Adults survival rate: {adults_survival_rate:.2f}%")
```

```python
elders = new_df[new_df["Age"] > 50]
elders_survival_rate = elders["Survived"].mean() * 100
print(f"Elders Survival Rate: {elders_survival_rate:.2f}%")
```
